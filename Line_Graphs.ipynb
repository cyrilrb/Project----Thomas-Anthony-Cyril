```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('game_highs_stats.csv')

df.head()
```

```python
df = df.set_index('Player')
```

```python
rs_df = df.loc[df['RSorPO'] == 'Regular Season',:]
po_df = df.loc[df['RSorPO'] == 'Playoffs',:]
LBJ_RG = ['S1','S2','S3','S4','S5','S6','S7','S8','S9','S10','S11','S12','S13','S14','S15','S16']
LBJ_PO = ['S3','S4','S5','S6','S7','S8','S9','S10','S11','S12','S13','S14','S15']
MJ_RG = ['S1','S2','S3','S4','S5','S6','S7','S8','S9','S10','S11','S12','S13','S14','S15']
MJ_PO = ['S1','S2','S3','S4','S5','S6','S7','S8','S9','S10','S11','S12','S13']
KB_RG = ['S1','S2','S3','S4','S5','S6','S7','S8','S9','S10','S11','S12','S13','S14','S15','S16','S17','S18','S19','S20']
KB_PO = ['S1','S2','S3','S4','S5','S6','S7','S8','S10','S11','S12','S13','S14','S15','S16']
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3,figsize=(30,5))

ax1.plot(LBJ_RG,rs_df.loc['Lebron James']['PTS'], marker = 'o', label= 'Regular Season')
ax1.plot(LBJ_PO ,po_df.loc['Lebron James']['PTS'], marker = 'o', label= 'Playoffs')
ax1.set_title('Lebron Game High Points')
ax1.set_ylim(20,85)
ax1.legend()
ax2.plot(MJ_RG,rs_df.loc['Michael Jordan']['PTS'], marker = 'o', label= 'Regular Season')
ax2.plot(MJ_PO,po_df.loc['Michael Jordan']['PTS'], marker = 'o', label= 'Playoffs')
ax2.set_title('Jordan Game High Points')
ax2.set_ylim(20,85)
ax2.legend()
ax3.plot(KB_RG,rs_df.loc['Kobe Bryant']['PTS'], marker = 'o', label= 'Regular Season')
ax3.plot(KB_PO,po_df.loc['Kobe Bryant']['PTS'], marker = 'o', label= 'Playoffs')
ax3.set_title('Kobe Game High Points')
ax3.set_ylim(20,85)
ax3.legend()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3,figsize=(30,5))

ax1.plot(LBJ_RG,rs_df.loc['Lebron James']['AST'], marker = 'o', label= 'Regular Season')
ax1.plot(LBJ_PO ,po_df.loc['Lebron James']['AST'], marker = 'o', label= 'Playoffs')
ax1.set_title('Lebron Game High Assists')
ax1.set_ylim(0,20)
ax1.legend()
ax2.plot(MJ_RG,rs_df.loc['Michael Jordan']['AST'], marker = 'o', label= 'Regular Season')
ax2.plot(MJ_PO,po_df.loc['Michael Jordan']['AST'], marker = 'o', label= 'Playoffs')
ax2.set_title('Jordan Game High Assists')
ax2.set_ylim(0,20)
ax2.legend()
ax3.plot(KB_RG,rs_df.loc['Kobe Bryant']['AST'], marker = 'o', label= 'Regular Season')
ax3.plot(KB_PO,po_df.loc['Kobe Bryant']['AST'], marker = 'o', label= 'Playoffs')
ax3.set_title('Kobe Game High Assists')
ax3.set_ylim(0,20)
ax3.legend()
```
