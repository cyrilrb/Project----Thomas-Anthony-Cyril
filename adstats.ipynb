```python
# libraries
import numpy as np
import matplotlib.pyplot as plt

fig1, ax1 = plt.subplots(figsize=(30, 10))
 
# set width of bars
barWidth = 0.25
 
# set heights of bars
bars1 = [31.7, 31.6, 31.2, 31.1, 29.8, 29.7, 29.4, 27.8, 27.7, 27.5, 25.8, 25.2, 22.1, 20.7, 19.3, 0, 0, 0, 0, 0]
bars2 = [18.3, 25.7, 28.1, 24.5, 29.1, 31.7, 31.1, 27.3, 30.7, 31.6, 29.3, 25.9, 27.5, 27.0, 28.6, 25.6, 25.5, 24.7, 0, 0]
bars3 = [14.4, 18.5, 18.9, 21.7, 24.5, 23.2, 26.2, 23.7, 23.3, 28, 26.1, 24.2, 24.4, 21.9, 23.9, 21.9, 23, 10.7, 17.6, 14.9]
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
# Make the plot
plt.bar(r1, bars1, color='r', width=barWidth, edgecolor='white', label='Jordan')
plt.bar(r2, bars2, color='y', width=barWidth, edgecolor='white', label='Lebron')
plt.bar(r3, bars3, color='m', width=barWidth, edgecolor='white', label='Kobe')
 
# Add xticks on the middle of the group bars
plt.xlabel('Seasons in NBA', fontsize = 25)
plt.xticks([r + barWidth for r in range(len(bars1))], ['S1', 'S2', 'S3', 'S4', 'S5','S6', 'S7', 'S8', 'S9', 'S10','S11', 'S12', 'S13', 'S14', 'S15','S16', 'S17', 'S18', 'S19', 'S20'], fontsize= 25)
plt.ylabel ('PER', fontsize= 25)
plt.title("Player Efficiency Rating (Regular Season)", fontsize=25, fontweight='bold')
plt.yticks(fontsize= 25)
plt.axhline(y=15)


# Create legend & Show graphic
plt.legend()
plt.show()
```
